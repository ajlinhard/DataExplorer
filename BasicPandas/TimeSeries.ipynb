```python
df_time = pd.DataFrame()
df_time['time_scale'] = pd.date_range('2025-01-01', periods=10, freq='D')
# df_time['time_scale'] = pd.concat([pd.date_range('2025-01-01', periods=10, freq='D'), pd.DataFrame([np.nan])], axis=0)
print(df_time.shape)
df_time.head(15)
```

```python
df_time.columns
print(df_time.index)
df_time.dtypes
```

```python
df_nan = pd.DataFrame([np.nan, 1], columns=['time_scale'])
df_time_comb = pd.concat([df_time, df_nan], axis=0)

df_time_comb.head(15)
# print(df_time_comb.shape)
```

```python
df_time_wide = pd.concat([df_time, df_nan], axis=1)
df_time_wide.head(15)
```

```python
df_time_wide.rename(columns={0: 'time_scale_1'}, inplace=True)
df_time_wide.rename(mapper=str.upper, axis=1, inplace=True)
df_time_wide.head(15)
```

```python
df_time_wide.columns = ['time_scale', 'time_scale2']
l_col = df_time_wide.columns.tolist()
l_col[1] = 'renamed_time_scale'
df_time_wide.columns = l_col
df_time_wide.head(15)
```

```python
l_index = df_time_wide.index.tolist()
l_index[0] = 'first_row'
df_time_wide.index = l_index
df_time_wide.head(15)
```

```python
df_time_wide.iloc[0::2].head(10)

df_filter = (df_time_wide['time_scale'].dt.year == 2025) & (df_time_wide['renamed_time_scale'].notna())
df_filter.head(10)
```

```python
df_time_wide[df_filter].head(10)
```

```python
df_time[['year', 'month', 'day']] = df_time['time_scale'].dt.year, df_time['time_scale'].dt.month, df_time['time_scale'].dt.day
df_time.head(15)
```

```python
import random
df_time['time_response'] = df_time['time_scale'].apply(lambda x: x + pd.Timedelta(days=random.randint(0, 10)))
```

```python
conditions = [df_time['time_scale'].dt.dayofweek < 5, 
    df_time['time_scale'].dt.dayofweek == 5, 
    df_time['time_scale'].dt.dayofweek == 6
    ]
choices = ['weekday', 'sat', 'sun']
default_choice = 'unknown'
df_time['time_category'] = np.select(conditions, choices, default=default_choice)
df_time.head(15)
```
