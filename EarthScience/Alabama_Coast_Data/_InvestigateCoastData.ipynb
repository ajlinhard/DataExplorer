## Aggregations

```python
df_csv_agg = df_csv.groupby('CountryRegionCode').agg({
    'GeographyKey' : [('GeographyCount', 'count'), ('GeographyDistCount', 'nunique')],
    'PostalCode' : [('PostalMin', 'min'), ('PostalMax', 'max')],
})
df_csv_agg.head(10)
```
