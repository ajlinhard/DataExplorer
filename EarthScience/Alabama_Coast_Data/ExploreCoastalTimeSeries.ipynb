```python
import pandas as pd
import numpy as np

# When loading a zipped CSV file, you can specify the compression type, but leave out the underlying file name.
df_csv = pd.read_csv(r"C:\Users\dalej\OneDrive\Documents\Data\hydrodata\di_met_min.2023.csv.zip", 
    compression='zip', 
    encoding='utf-8', 
    low_memory=False)
df_csv.head(10)
```

```python
df_csv['ts'] = pd.to_datetime(df_csv['ts'], format='%Y-%m-%d %H:%M:%S+00', errors='coerce')
```

```python
df_csv.head(5)
```

```python
df_csv.sort_values(by='ts', inplace=True)
df_csv['order'] = np.arange(len(df_csv)) + 1
df_csv['order_pd'] = df_csv.reset_index().index + 1
```

```python
df_csv['airtemp_next'] = df_csv['airtemp1'].shift(-1)
df_csv['order_lead'] = df_csv['order'].shift(-1)
df_csv['order_lag'] = df_csv['order'].shift(1).fillna(0)
```

```python

df_csv[['ts', 'airtemp1', 'airtemp_next', 'order', 'order_pd', 'order_lead', 'order_lag']].head(10)
```

```python
df_csv['airtemp_dph'] = (df_csv['airtemp1'] - df_csv['airtemp_next'])
```

```python
df_csv[['ts', 'airtemp1', 'airtemp_next', 'order', 'order_pd', 'order_lead', 'order_lag', 'airtemp_dph']].head(10)
```

```python
df_csv.describe()
```

```python
air_temp_bin = [-1000.0, -5.00, -1.00, -0.25, -.01, .01, 0.25, 1.00, 5.00, 1000.0]
air_labels = ['n super critical', 'n critical', 'n high', 'n moderate', 'neutral', 'moderate', 'high', 'critical', 'super critical']
df_csv['change_level'] = pd.cut(df_csv['airtemp_dph'], bins=air_temp_bin, labels=air_labels)
df_csv[['ts', 'airtemp1', 'airtemp_next', 'order', 'order_pd', 'order_lead', 'order_lag', 'airtemp_dph', 'change_level']].head(10)
df_csv['change_level'].value_counts()
```
