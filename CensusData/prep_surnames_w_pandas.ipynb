```python
def mixed_data_type(df: pd.DataFrame) -> pd.DataFrame:
    """
    Convert all columns to string type, except for 'rank' which should remain as int.
    """
    df_type = pd.DataFrame()
    for col in df.columns:
        df_type[col] = df[col].apply(type).apply(str)
    return df_type
```

```python
df_types = mixed_data_type(df_census_surname)
df_types.head(10)
```

```python
def is_numeric_string(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
    
for col, dtype in df_census_surname.dtypes.items():
    if dtype == 'object':
        b_result = ~df_census_surname[col].apply(is_numeric_string).all()
        if b_result:
            print(f"Column '{col}' contains non-numeric values.")
```

## Continue Rank and Transform

```python
df_ranked = df_census_surname[(df_types['name'] != "<class 'float'>") & (df_census_surname['name'] != 'ALL OTHER NAMES')].sort_values(by='count', ascending=False)
df_ranked.head(10)
```

```python
df_ranked.reset_index(inplace=True, drop=True)
# Method 1: Using index as unique rank
df_ranked.index.name = 'unqiue_rank'
df_ranked.reset_index(inplace=True)
# Method 2: Using index for unique rank
# df_ranked['unqiue_rank'] = df_ranked.index + 1
df_ranked['first_letter'] = df_ranked['name'].str[0].str.upper()
df_ranked['letter_rank'] = df_ranked.groupby('first_letter').cumcount() + 1

# View Results
df_ranked[['name', 'rank', 'cum_prop100k', 'unqiue_rank', 'first_letter', 'letter_rank']].head()
```
